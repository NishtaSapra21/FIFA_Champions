```python
# Import Splinter, BeautifulSoup, and Pandas
from splinter import Browser
from bs4 import BeautifulSoup as soup
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
import requests
import lxml.html as lh
from parsel import Selector
```

```python
# Set the executable path and initialize Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the mars nasa news site
url = 'https://www.foxsports.com/soccer/2022-fifa-world-cup/history/'
browser.visit(url)
```

```python
# Create object page
page = requests.get(url).text
fifa_soup = soup(page, 'html.parser')
```

```python
html = browser.html
soup = soup(html, 'html.parser')
soup
```

```python
# Obtain information from tag <table>
fifa_header = fifa_soup.find('table', class_='sticky-header')
fifa_header
```

```python
# Obtain every title of columns with tag <th>
headers = []
for i in fifa_header.find_all('th'):
    title = i.text
    clean_title = "".join([ele.strip() for ele in title.split("\n")])
    if clean_title == '':
        clean_title = "Year"
    headers.append(clean_title)
headers
```

```python
fifa_df = pd.DataFrame(columns = headers)
fifa_df.columns
```

```python
# Obtain information from tag <table>
fifa_data = fifa_soup.find('tbody', class_='row-data')
#fifa_data
```

```python
 fifa_df
```

```python
# Create a for loop to fill fifa_df
for j in fifa_data.find_all('tr')[0:]:
    row_data = j.find_all('td')
    row = [i.text for i in row_data]
    length = len(fifa_df)
    fifa_df.loc[length] = row
fifa_df
```

```python
# Remove whitespace and newline from entire DataFrame
def remove_whitespace(dataFrame):
    try:
        # remove spaces inside and outside of string
        dataFrame = "".join([ele.strip() for ele in str(dataFrame).split("\n")])

    except:
        pass
    return dataFrame
```

```python
clean_fifa_df = fifa_df.applymap(remove_whitespace)
clean_fifa_df
```

```python
# Winners of FIFA 
clean_fifa_df["CHAMPION"].value_counts()
```

```python
clean_fifa_df.dtypes
```

```python
clean_fifa_df['TEAMS'] = clean_fifa_df['TEAMS'].astype(str).astype(int)
clean_fifa_df['MATCHES'] = clean_fifa_df['MATCHES'].astype(str).astype(int)
clean_fifa_df['GOALS'] = clean_fifa_df['GOALS'].astype(str).astype(int)
```

```python
max_goal = clean_fifa_df.loc[clean_fifa_df["GOALS"].idxmax()]
max_goal
```

```python
# Write clean CSV file
clean_fifa_df.to_csv("fifa_winners.csv", index=False)
```
